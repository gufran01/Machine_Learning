### Outlier Treament  
Using 1.5 IQR rule

```python
q3=df['person_age'].quantile(0.75)
q1=df['person_age'].quantile(0.25)
iqr=q3-q1
print("IQR : ",iqr)
lower=np.round(q1-1.5*(iqr))
print("Lower : ",lower)
upper=np.round(q3+1.5*(iqr))
print("Upper : ",upper)
```

```python
df[(df['person_age']>lower) & (df['person_age']<upper)]
```
